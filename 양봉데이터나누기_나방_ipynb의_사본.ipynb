```python
#jpg, txt 파일 짝 안맞을때 돌리는 코드
import os

data_dir = '/content/drive/MyDrive/bee_yolo_b'  # 데이터 디렉토리 경로

# 경로에서 .jpg 및 .txt 파일 가져오기
jpg_files = [f for f in os.listdir(data_dir) if f.endswith('.jpg')]
txt_files = [f for f in os.listdir(data_dir) if f.endswith('.txt')]

# .jpg 파일의 이름에서 확장자 제거 후 집합으로 변환
jpg_file_names = {os.path.splitext(f)[0] for f in jpg_files}

# .txt 파일의 이름에서 확장자 제거 후 집합으로 변환
txt_file_names = {os.path.splitext(f)[0] for f in txt_files}

# .jpg 파일에는 있지만 .txt 파일에는 없는 파일 출력
missing_txt_files = jpg_file_names - txt_file_names
for file_name in missing_txt_files:
    print(f".jpg 파일은 있지만 .txt 파일이 없음: {file_name}")

# .txt 파일에는 있지만 .jpg 파일에는 없는 파일 출력
missing_jpg_files = txt_file_names - jpg_file_names
for file_name in missing_jpg_files:
    print(f".txt 파일은 있지만 .jpg 파일이 없음: {file_name}")
```

```python
# 원본 데이터가 있는 디렉토리
data_dir = '/content/drive/MyDrive/bee_yolo_mmh2/data/data_total'

# 데이터를 저장할 디렉토리 생성
train_dir = '/content/drive/MyDrive/bee_yolo_mmh2/train'
test_dir = '/content/drive/MyDrive/bee_yolo_mmh2/test'
val_dir = '/content/drive/MyDrive/bee_yolo_mmh2/val'

os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
```

```python
# 모든 파일 가져오기
all_files = os.listdir(data_dir)

# 이미지 파일과 라벨 파일 분리
jpg_files = [f for f in all_files if f.endswith('.jpg')]
txt_files = [f for f in all_files if f.endswith('.txt')]

# 파일 이름을 기준으로 정렬
jpg_files.sort()
txt_files.sort()

# 이미지 파일과 라벨 파일을 짝으로 맞추어 복사
for jpg_file, txt_file in zip(jpg_files, txt_files):
    source_jpg = os.path.join(data_dir, jpg_file)
    source_txt = os.path.join(data_dir, txt_file)

    if random.random() < 0.7:
        destination_dir = train_dir
    elif random.random() < 0.9:
        destination_dir = test_dir
    else:
        destination_dir = val_dir

    destination_jpg = os.path.join(destination_dir, jpg_file)
    destination_txt = os.path.join(destination_dir, txt_file)

    shutil.copy(source_jpg, destination_jpg)
    shutil.copy(source_txt, destination_txt)

    print(f"복사 완료: {jpg_file} 및 {txt_file} -> {destination_dir}")
```
